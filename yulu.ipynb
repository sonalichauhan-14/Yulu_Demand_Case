# Model

```python
# import linear regression model library
from sklearn.linear_model import LinearRegression
```

```python
x = df[['humidity','temp','windspeed','season','holiday']]
y= df['count']
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# train model
model= LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# visualize the model 
#plt.figure(figsize=(10,8))
plt.scatter(y_test,y_pred,alpha=0.5,color='darkorange',label='Predicted vs. Actual')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],'r--',linewidth=2,label='Predict line')
plt.xlabel('Actual Demand')
plt.ylabel('Predicted Demand')
plt.title('Linear Regression: Actual vs. Predicted')
plt.legend()
plt.show()
```

The scatter plot shows that the linear regression model underpredicts bike rental demand, with predicted values consistently falling below actual values. This indicates limited model accuracy and suggests potential for improvement through feature tuning or alternative models.

**Evalution**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
print('mae:',mean_absolute_error(y_test,y_pred))
print('mse:',mean_squared_error(y_test,y_pred))
print('r2_score:',r2_score(y_test,y_pred))
```

# Case Study: Predicting Demand for Yulus Shared Electric Cycles

**Problem Statement**
Yulu, a leading micro-mobility provider in India, experienced a significant decline in revenue and sought to understand the factors influencing demand for its shared electric cycles. The objective was to build a predictive model that could guide strategic decisions related to fleet deployment, pricing, and zone placement.
  
  **Analytical Approach**
  
  Multiple regression models were tested, including Linear Regression, Lasso, Ridge, and Random Forest. Initial results showed limited predictive power (R  27%), indicating that overfitting was not the issue. Instead, the limitation stemmed from insufficient feature representation.
  
  **Feature Engineering**  
  
  To enhance model performance, several new features were engineered:
- Time-based variables: hour, weekday, month
- Interaction terms: temp  humidity, is_weekend
These additions captured temporal and behavioral patterns critical to rental demand.
  
  **Model Performance**
  
  After retraining with the enriched feature set:
- Random Forest achieved R = 86.4%
- MAE = 44.31
- MSE = 4,488.92
This marked a substantial improvement in predictive accuracy and model fit.
  
  **Key Insights**
  
  - Hour was the most influential feature, confirming that demand is highly time-sensitive
- Environmental factors such as temperature and humidity played a moderate role
- Business variables like workingday had more impact than holiday or is_weekend
  
  **Business Impact**
  
-The model provides actionable insights to:
- Optimize fleet distribution based on peak hours
- Refine zone placement strategies
- Align operational planning with real-time demand patterns
